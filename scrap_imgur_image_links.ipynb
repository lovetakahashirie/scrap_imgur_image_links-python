```python
from bs4 import BeautifulSoup
import requests
import time

from selenium.webdriver.chrome import options
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# PS: this is chrome version 109, so if ur chrome is using another version, go to download an updated one
path = 'C:\Program Files (x86)\chromedriver.exe'
url = 'https://imgur.com/a/yODQvDV'

driver = webdriver.Chrome(service=Service(path))
driver.get(url)

# waits maximum 3 seconds for element's presence. If it is found after 2 seconds then code execution will be continued without wait for more 1 seconds.
driver.implicitly_wait(3)

# use if because no everytime need to click

# click yes
if (driver.find_element(By.CSS_SELECTOR, '#root .WallPage .Wall .Wall-choices div').click() == True):
    pass

driver.implicitly_wait(3)

# choose load more
if (driver.find_element(By.CSS_SELECTOR, '#root .loadMoreContainer').click() == True):
    pass
```

```python
# after pressing that button, we need to go back to the top of the page, preparing for the scrap below(need to start at top)
driver.execute_script("window.scrollTo(0, 0)")
# u can also find any element then send keys(home), both ok, but u need to import Keys first

links = []
y = 700
loop =50  # cant think of a perfect way, but just choose a optimal number depending on the # of pics. fyi:86 is loop=70
# the reason for us to loop is that, the f12 will change according to where u scroll
# for example, if you are at the top of the page, and u parse the page_source to bs4, u can only get the first n < total pic links

# PS: dont scroll too much everytime, ow u will miss some pics

for j in range(loop):
    # seems that is easier to pass whole page source(but need scrolling) then extract links
    soup = BeautifulSoup(driver.page_source, 'lxml')
    preprocessed_links = soup.find_all('img', class_="image-placeholder")
    # links are in this form 'https://i.imgur.com/LRYQtIo_d.webp?maxwidth=760&fidelity=grand'
    # so i need to extract unitl _ and add jpeg
    # use find() because we only have one _

    for i in preprocessed_links:
        index = i['src'].find('_')
        links.append(i['src'][:index] + '.jpeg')
        links = list(dict.fromkeys(links))
    # driver.execute_script('window.scrollBy(0,1000)')
    driver.execute_script('window.scrollBy(0,' + str(y) + ')')
    # scroll_down()
    time.sleep(0.1)

# i put length here for u to check whether u are getting all the links
len(links), links
```

```python
# write list to txt file, with \n
# just the classic way
# save in cd
def write(txt_name):
    with open(txt_name + '.txt', 'w') as f:
        for i in links:
            f.write(f'{i}\n')
```
